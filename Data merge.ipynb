```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

df1 = pd.read_csv("ZippedData/tmdb.movies.csv", index_col=0) #converting the file to a dataframe
df1.head()#print out the first five lines
```

```python
df2 = pd.read_csv("ZippedData/title.basics.csv") #converting the file to a dataframe
df2.head()#print out the first five lines
```

```python
df3 = pd.read_csv("ZippedData/title.ratings.csv") #converting the file to a dataframe
df3.head()#print out the first five lines
```

```python
df4 = pd.read_csv("ZippedData/tn.movie_budgets.csv") #converting the file to a dataframe
df4.head()#print out the first five lines
```

```python
df5 = pd.read_csv("ZippedData/rt.movie_info.tsv", sep ='\t') #converting the file to a dataframe
df5.head()#print out the first five lines
```

```python
merged_df = pd.merge(df1, df2)# merging df1 and df2
merged_df.head()#printing the first five lines
merged_df.shape
```

```python
joined_df = merged_df.merge(df3, left_index=True, right_index=True, how='outer')
joined_df.head()
```

```python
joined_df2 = joined_df.merge(df4, left_index=True, right_index=True, how='outer')#performs another join operation using an outer join between the joined_df 
joined_df2.head()
```

```python
joined_df3 = joined_df2.merge(df5, left_index=True, right_index=True, how='outer')#performs another join operation using an outer join between the joined_df 
joined_df3.head()
```
